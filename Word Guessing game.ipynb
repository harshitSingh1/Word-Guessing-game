```python
import random
lst = open('sevenletterwords.txt', 'r').readlines()
for i in range(len(lst)):
    lst[i] = lst[i].replace("\n", "")
password = random.choices(lst)[0].upper()
GARBAGE_CHARS = '~!@#$%^&*()_+-={}[]|;:,.<>?/'
print("\033[1;37;40m WELCOME TO THE WORLD OF GUESSING \033[0;0m \n")
print('''You are given clues after each guess. For example, if the secret password is MONITOR but the
player guessed MONIXYZ, they are given the hint that coloured letters
were correct,by displaying the matched part in different colour. You get 5 guesses.\n 
NOTE: The password contains 7 capital letters!! \n''')

def main():
    flag = False
    input('Press Enter for hint...')
    x=hint()
    for i in range(5):
        guess = input(f'Try{i+1}: Guess the password: ')
        result = ""
        start = "\033[1;31m"
        end = "\033[0;0m"
        if len(guess) != len(password):
            print("\n The real password contain 7 letters but {} contain {} letters \n".format(guess,len(guess)))
            pass
        else:
            for j in range(len(password)):
                if guess[j] == password[j]:
                    result += start + guess[j] + end
                else:
                    result += guess[j]
            print("The red coloured letters are matching: ", result)
        if guess == password:
            print('Correct Password')
            flag = True
            break
        else:
            print(f'Incorrect Password try again,{4-i} tries remaining')
    if not(flag):
        print('Correct password is', password)
def hint():
    b = 0
    c = []
    while len(c)<10:
        a = 0
        for k in range(len(password)):
            if password[k].lower() == lst[b][k]:
                a+=1
            else:
                pass
        if a>2:
            c.append(lst[b].upper())
        b+=1
    for l in range(10):
        char = ''
        for j in range(40):
            char += random.choice(GARBAGE_CHARS)
        Index = random.randint(0, 30)
        Hint = (char[:Index] +"  "+ c[l] +"  "+ char[Index + 7:])
        print(Hint)
if __name__ == '__main__':
    try:
        main()
    except KeyboardInterrupt:
        print("ERROR")
```
